```python
"""
Created on Sat Oct 19 2019

@author: sharmaine.aguilar
"""

"""
DocString:

    A) A classic match-up style dating game for desparate singles (wonder who could be?). 
    Go through three rounds with different potential dates to see if you can win anyone of their hearts.
    
        Date Scenario 1: Cooking dinner
        Date Scenario 2: Spontaneous travel trip
        Date Scenario 3: Meet the parents/friends
    
    B) Assumptions:
        Two known genders (Female/Male)
    
    C) Known bugs/errors:
        Possibility of same partner candidate (due to random choice) after host asks to try different candidate.
        Limitations of input answers based on choices presented
        
    D) Sources:
        Based on Myers Briggs personality types INTJ, ESFP and ENFP
        https://www.bustle.com/p/the-ideal-date-idea-for-each-myers-briggs-type-8696656
        https://www.yourtango.com/2018316445/travel-destination-ideas-for-each-myers-briggs-personality-type
    
"""

partner1 = ["1",
            "will literally charm your socks off! Confidence and love for connection (not wifi, sorry) are two of the most attractive traits",
            "rome","gathering"]
partner2 = ["2",
            "pursues knowledge! Know your books, algorithms, and which companies are headed for IPO because intelligence is all the key you need",
            "edinburgh","help"]
partner3 = ["3","is all about putting on a good show, and brings the party to the room! Talk about fun",
            "buenos aires","wild"]

partner_list = [partner1, partner2, partner3]

from sys import exit
import random

def start():
    
    global contestant_name
    global pronoun
    
    print("""
    Calling all the single (and so ready to mingle) men and ladies!
    The night has just begun. We’re ready to match you up with our
    fiery candidates from the Bay Area.

    \t\tWelcome to…
    
    \t\t\tThe Perfect Match: Bay Area edition!""")
    
    input("\n<Press Enter to Continue> ")
    
    contestant_name = ""
    while contestant_name == "":
        contestant_name = input("""\nHost: Someone is feeling lucky tonight! Tell us, what\'s your name?: """)
        if contestant_name != "":
            print(f"""
            \nYou: Hi, everyone. My name is {contestant_name.capitalize()}, and I can\'t wait 
            to finally date someone so people won\'t think I\'m a lonely person...""")
            break
        else:
            print("Host: Hey, we know you have a name. Someone called you last night. Try again?")
 
    pronoun = ""
    while pronoun == "":
        print("""\nAre you looking to date...
              Women
              Men """)
    
        partner_gender = input("> ").lower()
        if partner_gender in ("woman","women"):
            pronoun = "she"
        elif partner_gender in ("man","men"):
            pronoun = "he"
        else:
            print("Host: Hmm...that's not really in any of our choices. Try again?")
    
    print(f"""
    \nHost: Great! We have a few amazing {partner_gender} lined-up tonight.
    I can tell you, it\'s gonna be exciting! Let\'s proceed shall we...
    
    Partners, please step up front for Round 1!
    """)
    
    input("\n<Press Enter to Continue>")
    
    date1_cooking()
    
def date1_cooking():
    
    global random_partner
    random_partner = random.choice(partner_list)     
    
    print(f"""
    \nHost: Partner {random_partner[0]} {random_partner[1]}. {pronoun.capitalize()} loves to cook and {pronoun} invited you over for dinner. You are thinking...
    """)
    
    print("""
        1) Anything Ghiradelli - Maybe, hot fudge sundae?
        2) Classic SF - clam chowder in sourdough bowl. That should do the trick!
        3) Vegan? Gluten-free? Keto? I can cook it all for you baby...  
    """)
    
    chances = 2
    
    from random import randint
    
    cooking_partner = randint(1,3)
    while chances > 0:
        cooking_answer = input(f"\nWhich one are you cooking?: ").lower()
        
        if cooking_answer in ("1", "ghiradelli","hot fudge", "hot fudge sundae","sundae","fudge"):
            if cooking_partner == 1:
                print(f"""\nYum! Partner {random_partner[0]} is full. As they say, a way to their hearts is through their stomachs.""")
                input("\n<Press Enter to Continue>")
                date2_trip()
                break
            else:
                chances -=1
                print(f"""Host: Not quite what Partner {random_partner[0]} is expecting. You only have {chances} chance left. Try again?""") if chances else 0
                
        elif cooking_answer in ("2", "clam chowder", "classic sf", "classic", "sf", "sourdough bowl"):
            if cooking_partner == 2:
                print(f"""\nYum! Partner {random_partner[0]} is full. As they say, a way to their hearts is through their stomachs.""")    
                input("\n<Press Enter to Continue>")
                date2_trip()
                break
            else:
                chances -=1
                print(f"""Host: Not quite what Partner {random_partner[0]} is expecting. You only have {chances} chance left. Try again?""") if chances else 0
        
        elif cooking_answer in ("3", "vegan", "gluten", "gluten-free", "keto", "all"):
            if cooking_partner == 3:
                print(f"""\nYum! Partner {random_partner[0]}. As they say, a way to their hearts is through their stomachs.""")        
                input("\n<Press Enter to Continue>")
                date2_trip()
                break
            else:
                chances -=1
                print(f"""Host: Not quite what Partner {random_partner[0]} is expecting. You only have {chances} chance left. Try again?""") if chances else 0
        
        else:
            print("Host: Hmm...We didn\'t quite get that. Try again?")
            
    fail()
    
def date2_trip():
        
    print("""
    Host: Okay, time for round 2. Now that we've got our dinners sorted, how about where should we go next?
    """)
    input("<Press Enter to Continue>")
    
    print(f"""
    Host: Based on Partner {random_partner[0]}\'s personality, you are thinking...
    
        1) Rome - perfect for those who are confident enough to build connections with the locals (competition, eh?)
        2) Endinburgh - learn about history, and more! Also, HP nerds will go crazy here.
        3) Buenos Aires - nightlife, drinks, dancing, repeat!
    """)
    
    chances_2 = 2
    
    while chances_2 > 0:
        trip_answer = input(f"Where are you booking your flights? ").lower()
            
        if trip_answer in ("1", "one", "rome", "italy", "confident", "connections"):
            trip_answer = 'rome'
            if trip_answer == random_partner[2]:
                print("Host: Ding, ding, ding! You got it. Wow, you are on fire!")
                input("\n<Press Enter to Continue>")
                date3_lovelang()
                break
            else:
                chances_2 -= 1
                print(f"Host: Yeah, not quite...You have {chances_2} left. Try again?") if chances_2 else 0
        
        elif trip_answer in ("2", "two", "edinburgh", "history","scotland", "nerds", "harry potter"):
            trip_answer = 'edinburgh'
            if trip_answer == random_partner[2]:
                print("Host: Ding, ding, ding! You got it. Wow, you are on fire!")
                input("\n<Press Enter to Continue>")
                date3_lovelang()
                break
            else:
                chances_2 -= 1
                print(f"Host: Yeah, not quite...You have {chances_2} left. Try again?") if chances_2 else 0
        
        elif trip_answer in ("3", "three", "buenos aires", "argentina", "party", "nightlife","drinks", "dancing"):
            trip_answer = 'buenos aires'
            if trip_answer == random_partner[2]:
                print("Host: Ding, ding, ding! You got it. Wow, you are on fire!")
                input("\n<Press Enter to Continue>")
                date3_lovelang()
                break
            else:
                chances_2 -= 1
                print(f"Host: Yeah, not quite...You have {chances_2} left. Try again?") if chances_2 else 0   
        else:
            print("Host: Hang on...We didn\'t quite get that. Try again?")
     
    fail()
    
def date3_lovelang():
        
    print("""
    Host: Finally, we get to the last round...Round 3! You only have one chance and by this time, you should
    have known your potential partner better!
    """)
    input("<Press Enter to Continue>")
    
    print(f"""
    Host: Everyone is different (spoiler alert?), and so do our love languages. To get boo\'s affection, you...
    
        1) Help with a project and offer advice
        2) Plan a surprise gathering with beloved plus deeply connected friends and family
        3) Support by being as wild (you do you - no judgment babe!) and extra playful ;)
    """)
    
    lovelang_answer = ""
    
    while lovelang_answer == "":
        lovelang_answer = input(f"How will you show love? ").lower()
        
        if lovelang_answer in ("1", "one", "help", "project", "help with a project", "help project", "offer advice" "advice"):
            lovelang_answer = 'help'
            if lovelang_answer == random_partner[3]:
                print("\nHost: OMG! This is it!")
                input("\n<Press Enter to Continue>")
                win()
                break
            else:
                fail()
        
        elif lovelang_answer in ("2", "two", "Plan", "surprise gathering", "surprise", "with beloved" "deeply connected", "friends","family"):
            lovelang_answer = 'gathering'
            if lovelang_answer == random_partner[3]:
                print("\nHost: OMG! This is it!")
                input("\n<Press Enter to Continue>")
                win()
                break
            else:
                fail()
        
        elif lovelang_answer in ("3", "three", "Support" "being as wild", "you do you", "no judgment", "extra playful"):
            lovelang_answer = 'wild'
            if lovelang_answer == random_partner[3]:
                print("\nHost: OMG! This is it!")
                input("\n<Press Enter to Continue>")
                win()
                break
            else:
                fail()
        else:
            print("Host: Hmm...We didn\'t quite get that.. Try again?")
            
    fail()
            
def win():
    print(f"""
    \nHost: Wow, {contestant_name.capitalize()}. Way to pay attention to all the details. 
    CONGRATULATIONS! You\'ve won the FIRST EVER Perfect Match: Bay Area Edition with Partner {random_partner[0]}!
    You\'ve finally graduated from singlehood! 
    
    That\'s all folks! We\'ll see you next time!
    
    """)
    input("\n<Press Enter to Exit> ")
    
    exit(0)
              
def fail():
    print(f"""\nHost: Woops. Looks like your date backed out. We thought {pronoun}\'s unique,
    but {pronoun} actually meant {pronoun}\'s basic...You deserve better next time.
    """)
    
    retry = ""
    
    while retry == "":
        retry = input(f"\n Would you like to try a different partner and go back to Round 1? (Yes/No)").lower()
        
        if retry == 'yes':
            date1_cooking()
        elif retry == 'no':
            print("\nHost: Thanks for playing! We\'ll see you next time!")
            exit(0)
        else:
            print("Host: Hmm...that's not really in any of our choices. Try again?")

start()
```
